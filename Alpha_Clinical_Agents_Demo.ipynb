## 2. Import Required Modules

```python
from agents.protocol_analyzer import ProtocolAnalyzer
from agents.section_writer import SectionWriter
from agents.meta_validator import MetaValidator
from agents.conflict_resolver import ConflictResolver
from agents.risk_predictor import RiskPredictor
from core.orchestrator import ClinicalOrchestrator

print("✅ All modules imported successfully")
```

## 3. Demo: Protocol Analysis

```python
# Sample clinical protocol
protocol = '''
PHASE III RANDOMIZED, DOUBLE-BLIND, PLACEBO-CONTROLLED STUDY

STUDY TITLE:
A Study of Investigational Drug vs Placebo in Patients with
Metastatic Non-Small Cell Lung Cancer (NSCLC)

STUDY POPULATION:
Planned enrollment: 600 patients

INCLUSION CRITERIA:
1. Histologically confirmed Stage IV NSCLC
2. ECOG performance status 0-1
3. Adequate organ function

PRIMARY ENDPOINT:
Overall Survival (OS)

SECONDARY ENDPOINTS:
- Progression-Free Survival (PFS)
- Objective Response Rate (ORR)
- Safety and tolerability
'''

# Analyze protocol
analyzer = ProtocolAnalyzer()
analysis = analyzer.execute({"protocol_text": protocol})

print("📊 Protocol Analysis Results:")
print(f"  Phase: {analysis['study_design']['phase']}")
print(f"  Planned Enrollment: {analysis['population']['planned_enrollment']} patients")
print(f"  Primary Endpoint: {analysis['endpoints']['primary']}")
print(f"  Confidence: {analysis['extraction_confidence']:.1%}")
```

## 4. Demo: Risk Prediction

```python
# Predict complexity before processing
risk_predictor = RiskPredictor()
risk = risk_predictor.execute({"protocol_text": protocol})

print("⚠️ Risk Prediction:")
print(f"  Complexity Score: {risk['complexity_score']}/10")
print(f"  Risk Level: {risk['risk_level']}")
print(f"  Recommended QA: {risk['recommended_qa_level']}")
print(f"  Est. Processing Time: {risk['estimated_time_hours']} hours")
```

## 5. Demo: Section Generation

```python
# Generate CSR section
writer = SectionWriter()
section = writer.execute({
    "protocol_structure": analysis,
    "section_name": "Methods"
})

print("📝 Generated Section:")
print(f"  Word Count: {section['word_count']}")
print(f"  Confidence: {section['confidence']:.1%}")
print(f"  Citations: {len(section.get('citations', []))}")
print("\n--- Preview ---")
print(section['draft_text'][:500] + "...")
```

## 6. Demo: Multi-Agent Validation

```python
# Validate generated section
validator = MetaValidator()
validation = validator.execute({
    "agent_outputs": {
        "ProtocolAnalyzer": analysis,
        "SectionWriter": {"sections": {"Methods": section}}
    },
    "sections": {"Methods": {"draft": section['draft_text']}}
})

print("✅ Validation Results:")
print(f"  QA Score: {validation['qa_score']}/100")
print(f"  Issues Found: {validation['total_issues']}")
print(f"  Auto-Corrected: {validation['auto_corrected']}")
print(f"  Escalated to Human: {validation['escalated_to_human']}")
```

## 7. Demo: Conflict Resolution

```python
# Simulate conflicting claims
conflict_resolver = ConflictResolver()

resolution = conflict_resolver.execute({
    "conflicts": [
        {
            "claim_a": {
                "text": "600 patients enrolled",
                "source": "protocol",
                "confidence": 0.95
            },
            "claim_b": {
                "text": "580 patients enrolled",
                "source": "statistical_report",
                "confidence": 0.98
            }
        }
    ]
})

print("🔧 Conflict Resolution:")
print(f"  Conflicts Resolved: {resolution['conflicts_resolved']}")
print(f"  Resolution Strategy: {resolution['resolution_strategy']}")

for r in resolution['resolutions']:
    print(f"\n  Winner: {r['winner_source']}")
    print(f"  Rationale: {r['rationale']}")
```

## 8. Full Pipeline Demo

```python
# Run full 3-agent pipeline
orchestrator = ClinicalOrchestrator()

# Register agents
orchestrator.register_agent(ProtocolAnalyzer())
orchestrator.register_agent(SectionWriter())
orchestrator.register_agent(MetaValidator())

print("🚀 Running Full Pipeline...")
print("=" * 60)

# Execute workflow
import time
start = time.time()

# Step 1: Analyze
analysis = orchestrator.send_message_sync({
    "sender": "user",
    "recipient": "protocol_analyzer",
    "payload": {"protocol_text": protocol}
})
print(f"✅ Step 1: Protocol analyzed")

# Step 2: Generate
section = orchestrator.send_message_sync({
    "sender": "protocol_analyzer",
    "recipient": "section_writer",
    "payload": {
        "protocol_structure": analysis.payload,
        "section_name": "Methods"
    }
})
print(f"✅ Step 2: Section generated")

# Step 3: Validate
validation = orchestrator.send_message_sync({
    "sender": "section_writer",
    "recipient": "meta_validator",
    "payload": {
        "agent_outputs": {"analysis": analysis.payload, "section": section.payload},
        "sections": {"Methods": section.payload}
    }
})
print(f"✅ Step 3: Validation complete")

elapsed = time.time() - start
print("\n" + "=" * 60)
print(f"⏱️ Total time: {elapsed:.2f}s")
print(f"🎯 QA Score: {validation.payload.get('qa_score', 'N/A')}")
```
